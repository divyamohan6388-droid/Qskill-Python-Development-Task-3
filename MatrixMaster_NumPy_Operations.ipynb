## 📦 Step 1: Import Libraries

```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print('✅ NumPy imported successfully!')
print(f'   NumPy version : {np.__version__}')
print()
print('━'*55)
print('   🔢  MatrixMaster: NumPy Operations Console')
print('━'*55)
```

## 🛠️ Step 2: Core Utility Functions

```python
# ─────────────────────────────────────────────────────────────────
#  DISPLAY HELPERS
# ─────────────────────────────────────────────────────────────────

def divider(char='━', width=55):
    print(char * width)

def section(title):
    print()
    divider()
    print(f'  {title}')
    divider()

def format_matrix(M, name='Matrix', decimals=4):
    """Pretty-print a NumPy matrix with borders."""
    rows, cols = M.shape
    col_widths = []
    for c in range(cols):
        w = max(len(f'{M[r,c]:.{decimals}g}') for r in range(rows))
        col_widths.append(max(w, len(f'Col{c+1}')))

    total_w = sum(col_widths) + 3*cols + 1
    header  = '  ' + name + f'  [{rows}×{cols}]'
    print()
    print(header)
    print('  ┌' + '─'*(total_w-2) + '┐')

    # Column header
    hdr = '  │ ' + '  '.join(f'C{c+1}'.center(col_widths[c]) for c in range(cols)) + ' │'
    print(hdr)
    print('  ├' + '─'*(total_w-2) + '┤')

    for r in range(rows):
        row_str = '  │ '
        for c in range(cols):
            val = M[r, c]
            cell = f'{val:.{decimals}g}'.rjust(col_widths[c])
            row_str += cell + '  '
        row_str = row_str.rstrip() + ' │'
        print(row_str)

    print('  └' + '─'*(total_w-2) + '┘')

def format_vector(v, name='Vector'):
    """Pretty-print a 1D array."""
    print(f'\n  {name}:')
    print('  [ ' + '  '.join(f'{x:.4g}' for x in v) + ' ]')

print('✅ Display utilities ready.')
```

## 📥 Step 3: Matrix Input Function

```python
def input_matrix(name='Matrix'):
    """
    Interactively read a matrix from the user.
    The user enters rows one at a time, space-separated.
    Type 'done' after the last row.
    """
    print(f'\n  📥 Enter {name} row by row (space-separated values).')
    print(  '     Type  "done"  after the last row.')
    print(  '     Example row → 1 2 3')
    divider('-', 45)

    rows = []
    row_num = 1
    while True:
        raw = input(f'    Row {row_num}: ').strip()
        if raw.lower() == 'done':
            if not rows:
                print('  ⚠️  No rows entered. Please enter at least one row.')
                continue
            break
        if not raw:
            print('  ⚠️  Empty input — enter numbers or type "done".')
            continue
        try:
            vals = [float(x) for x in raw.split()]
        except ValueError:
            print('  ❌ Invalid input — use numbers only (e.g. 1 2 3).')
            continue
        if rows and len(vals) != len(rows[0]):
            print(f'  ❌ Row length mismatch — expected {len(rows[0])} values, got {len(vals)}.')
            continue
        rows.append(vals)
        row_num += 1

    M = np.array(rows)
    format_matrix(M, name)
    return M

print('✅ Matrix input function ready.')
```

## ⚙️ Step 4: All Matrix Operation Functions (with Step-by-Step Working)

```python
# ═══════════════════════════════════════════════════════════════
#  1. ADDITION
# ═══════════════════════════════════════════════════════════════
def op_addition():
    section('➕  MATRIX ADDITION  (A + B)')
    print('  📖 Formula : C[i,j] = A[i,j] + B[i,j]')
    print('  📌 Rule    : Both matrices must have the SAME shape.\n')
    A = input_matrix('Matrix A')
    B = input_matrix('Matrix B')
    if A.shape != B.shape:
        print(f'\n  ❌ Shape mismatch: A{A.shape} ≠ B{B.shape}. Cannot add.')
        return
    C = np.add(A, B)
    print('\n  🔍 Step-by-Step Working:')
    rows, cols = A.shape
    for i in range(rows):
        for j in range(cols):
            print(f'     C[{i+1},{j+1}] = A[{i+1},{j+1}] + B[{i+1},{j+1}]'
                  f' = {A[i,j]:.4g} + {B[i,j]:.4g} = {C[i,j]:.4g}')
    format_matrix(C, 'Result  C = A + B')


# ═══════════════════════════════════════════════════════════════
#  2. SUBTRACTION
# ═══════════════════════════════════════════════════════════════
def op_subtraction():
    section('➖  MATRIX SUBTRACTION  (A − B)')
    print('  📖 Formula : C[i,j] = A[i,j] − B[i,j]')
    print('  📌 Rule    : Both matrices must have the SAME shape.\n')
    A = input_matrix('Matrix A')
    B = input_matrix('Matrix B')
    if A.shape != B.shape:
        print(f'\n  ❌ Shape mismatch: A{A.shape} ≠ B{B.shape}. Cannot subtract.')
        return
    C = np.subtract(A, B)
    print('\n  🔍 Step-by-Step Working:')
    rows, cols = A.shape
    for i in range(rows):
        for j in range(cols):
            print(f'     C[{i+1},{j+1}] = A[{i+1},{j+1}] − B[{i+1},{j+1}]'
                  f' = {A[i,j]:.4g} − {B[i,j]:.4g} = {C[i,j]:.4g}')
    format_matrix(C, 'Result  C = A − B')


# ═══════════════════════════════════════════════════════════════
#  3. MULTIPLICATION
# ═══════════════════════════════════════════════════════════════
def op_multiplication():
    section('✖️   MATRIX MULTIPLICATION  (A × B)')
    print('  📖 Formula : C[i,j] = Σ  A[i,k] × B[k,j]')
    print('  📌 Rule    : Columns of A must equal Rows of B.\n')
    A = input_matrix('Matrix A')
    B = input_matrix('Matrix B')
    if A.shape[1] != B.shape[0]:
        print(f'\n  ❌ Shape error: A is {A.shape}, B is {B.shape}.')
        print(f'     Need A-cols({A.shape[1]}) == B-rows({B.shape[0]}).')
        return
    C = np.matmul(A, B)
    print('\n  🔍 Step-by-Step Working (dot product per cell):')
    rows_a, cols_a = A.shape
    _, cols_b      = B.shape
    for i in range(rows_a):
        for j in range(cols_b):
            terms = ' + '.join(f'{A[i,k]:.4g}×{B[k,j]:.4g}' for k in range(cols_a))
            print(f'     C[{i+1},{j+1}] = {terms} = {C[i,j]:.4g}')
    format_matrix(C, 'Result  C = A × B')


# ═══════════════════════════════════════════════════════════════
#  4. TRANSPOSE
# ═══════════════════════════════════════════════════════════════
def op_transpose():
    section('🔄  MATRIX TRANSPOSE  (Aᵀ)')
    print('  📖 Formula : Aᵀ[i,j] = A[j,i]')
    print('  📌 Rule    : Rows become columns, columns become rows.\n')
    A = input_matrix('Matrix A')
    T = A.T
    print(f'\n  🔍 Step-by-Step Working:')
    print(f'     Original shape : {A.shape[0]}×{A.shape[1]}')
    print(f'     Transposed shape: {T.shape[0]}×{T.shape[1]}')
    rows, cols = A.shape
    for i in range(rows):
        for j in range(cols):
            print(f'     Aᵀ[{j+1},{i+1}] ← A[{i+1},{j+1}] = {A[i,j]:.4g}')
    format_matrix(T, 'Result  Aᵀ (Transpose)')


# ═══════════════════════════════════════════════════════════════
#  5. DETERMINANT
# ═══════════════════════════════════════════════════════════════
def op_determinant():
    section('📐  DETERMINANT  (det(A))')
    print('  📖 For 2×2: det = ad − bc')
    print('  📖 For n×n: Computed via LU decomposition (NumPy).')
    print('  📌 Rule    : Matrix must be SQUARE (n×n).\n')
    A = input_matrix('Matrix A (square)')
    r, c = A.shape
    if r != c:
        print(f'\n  ❌ Not square: {A.shape}. Determinant requires n×n matrix.')
        return
    det = np.linalg.det(A)
    print(f'\n  🔍 Step-by-Step Working:')
    if r == 2:
        a,b,c_,d = A[0,0], A[0,1], A[1,0], A[1,1]
        print(f'     det(A) = (a×d) − (b×c)')
        print(f'            = ({a:.4g}×{d:.4g}) − ({b:.4g}×{c_:.4g})')
        print(f'            = {a*d:.4g} − {b*c_:.4g}')
        print(f'            = {det:.6g}')
    elif r == 3:
        print(f'     Using Sarrus Rule / cofactor expansion along Row 1:')
        for j in range(3):
            minor = np.delete(np.delete(A, 0, axis=0), j, axis=1)
            cofac = ((-1)**j) * np.linalg.det(minor)
            print(f'     Cofactor C[1,{j+1}] = {(-1)**j}×det(M{j+1}) = {cofac:.4g}')
        print(f'     det(A) = Σ A[1,j]×C[1,j] = {det:.6g}')
    else:
        print(f'     Using NumPy LU decomposition for {r}×{r} matrix.')
        print(f'     det(A) = {det:.6g}')
    print()
    divider('-', 45)
    print(f'  🎯 det(A) = {det:.6g}')
    if abs(det) < 1e-10:
        print('  ⚠️  Determinant ≈ 0  →  Matrix is SINGULAR (no inverse).')
    else:
        print('  ✅  Matrix is NON-SINGULAR (invertible).')


# ═══════════════════════════════════════════════════════════════
#  6. INVERSE
# ═══════════════════════════════════════════════════════════════
def op_inverse():
    section('🔃  MATRIX INVERSE  (A⁻¹)')
    print('  📖 Property: A × A⁻¹ = A⁻¹ × A = I (Identity)')
    print('  📖 Formula : A⁻¹ = (1/det(A)) × Adjugate(A)')
    print('  📌 Rule    : Matrix must be SQUARE and NON-SINGULAR (det ≠ 0).\n')
    A = input_matrix('Matrix A (square)')
    r, c = A.shape
    if r != c:
        print(f'\n  ❌ Not square: {A.shape}.')
        return
    det = np.linalg.det(A)
    print(f'\n  🔍 Step-by-Step Working:')
    print(f'     Step 1 → Calculate det(A) = {det:.6g}')
    if abs(det) < 1e-10:
        print('  ❌ det(A) ≈ 0  →  Matrix is SINGULAR. Inverse does not exist.')
        return
    print(f'     Step 2 → det(A) ≠ 0, so inverse exists.')
    if r == 2:
        a,b,c_,d = A[0,0], A[0,1], A[1,0], A[1,1]
        print(f'     Step 3 → For 2×2: A⁻¹ = (1/{det:.4g}) × [[{d:.4g}, {-b:.4g}], [{-c_:.4g}, {a:.4g}]]')
    else:
        print(f'     Step 3 → Computing via Gauss-Jordan elimination (NumPy linalg.inv).')
    inv_A = np.linalg.inv(A)
    print(f'     Step 4 → Verify: A × A⁻¹ ≈ Identity matrix')
    identity_check = np.allclose(np.matmul(A, inv_A), np.eye(r))
    print(f'              Check passed: {identity_check} ✅' if identity_check
          else f'              Check FAILED ❌')
    format_matrix(inv_A, 'Result  A⁻¹ (Inverse)')
    print()
    print('  🔁 Verification  A × A⁻¹ =')
    format_matrix(np.round(np.matmul(A, inv_A), 6), 'A × A⁻¹  (should be Identity)')


# ═══════════════════════════════════════════════════════════════
#  7. EIGENVALUES & EIGENVECTORS
# ═══════════════════════════════════════════════════════════════
def op_eigen():
    section('🌀  EIGENVALUES & EIGENVECTORS')
    print('  📖 Definition: A × v = λ × v')
    print('     where λ = eigenvalue, v = eigenvector')
    print('  📖 Computed from characteristic equation: det(A − λI) = 0')
    print('  📌 Rule    : Matrix must be SQUARE.\n')
    A = input_matrix('Matrix A (square)')
    r, c = A.shape
    if r != c:
        print(f'\n  ❌ Not square: {A.shape}.')
        return
    eigenvalues, eigenvectors = np.linalg.eig(A)
    print(f'\n  🔍 Step-by-Step Working:')
    print(f'     Step 1 → Solve characteristic equation: det(A − λI) = 0')
    print(f'     Step 2 → Roots of the polynomial give eigenvalues (λ).')
    print(f'     Step 3 → For each λ, solve (A − λI)v = 0 for eigenvector v.')
    print()
    divider('-', 45)
    print(f'  📊 Results for {r}×{r} Matrix:')
    divider('-', 45)
    for i, (lam, vec) in enumerate(zip(eigenvalues, eigenvectors.T)):
        lam_str = f'{lam.real:.4g}' if abs(lam.imag) < 1e-10 else f'{lam:.4g}'
        print(f'\n  Eigenvalue  λ{i+1} = {lam_str}')
        vec_real = vec.real if np.allclose(vec.imag, 0) else vec
        vstr = '  '.join(f'{v:.4g}' for v in vec_real)
        print(f'  Eigenvector v{i+1} = [ {vstr} ]')
        # Verify A·v = λ·v
        Av   = np.matmul(A, vec.real)
        lv   = lam.real * vec.real
        check = np.allclose(Av, lv, atol=1e-6)
        print(f'  Verify A·v{i+1} ≈ λ{i+1}·v{i+1}: {"✅ PASS" if check else "❌ FAIL"}')


# ═══════════════════════════════════════════════════════════════
#  8. RANK
# ═══════════════════════════════════════════════════════════════
def op_rank():
    section('📊  MATRIX RANK  (rank(A))')
    print('  📖 Definition: Number of linearly independent rows/columns.')
    print('  📖 Method    : Computed via Singular Value Decomposition (SVD).')
    print('  📌 Rule    : Works for ANY matrix shape.\n')
    A = input_matrix('Matrix A')
    r, c = A.shape
    rank = np.linalg.matrix_rank(A)
    _, s, _ = np.linalg.svd(A)
    print(f'\n  🔍 Step-by-Step Working:')
    print(f'     Step 1 → Perform SVD: A = U × Σ × Vᵀ')
    print(f'     Step 2 → Singular values (diagonal of Σ):')
    for i, sv in enumerate(s):
        is_nonzero = abs(sv) > 1e-10
        print(f'              σ{i+1} = {sv:.4g}  →  {"Non-zero (counted)" if is_nonzero else "≈ 0 (not counted)"}')
    print(f'     Step 3 → Rank = number of non-zero singular values')
    print()
    divider('-', 45)
    print(f'  🎯 rank(A) = {rank}')
    print(f'     Matrix size     : {r}×{c}')
    print(f'     Max possible rank: min({r},{c}) = {min(r,c)}')
    if rank == min(r, c):
        print('  ✅ Matrix has FULL RANK.')
    else:
        print(f'  ⚠️  Matrix is RANK-DEFICIENT (rank < min dimensions).')


# ═══════════════════════════════════════════════════════════════
#  9. TRACE
# ═══════════════════════════════════════════════════════════════
def op_trace():
    section('➰  MATRIX TRACE  (tr(A))')
    print('  📖 Definition: Sum of all diagonal elements of a square matrix.')
    print('  📖 Formula   : tr(A) = A[0,0] + A[1,1] + ... + A[n,n]')
    print('  📌 Property  : tr(A) = sum of all eigenvalues of A.')
    print('  📌 Rule      : Matrix must be SQUARE.\n')
    A = input_matrix('Matrix A (square)')
    r, c = A.shape
    if r != c:
        print(f'\n  ❌ Not square: {A.shape}. Trace requires n×n matrix.')
        return
    trace = np.trace(A)
    print(f'\n  🔍 Step-by-Step Working:')
    print(f'     Diagonal elements of {r}×{r} matrix:')
    diag_parts = []
    for i in range(r):
        print(f'     A[{i+1},{i+1}] = {A[i,i]:.4g}')
        diag_parts.append(f'{A[i,i]:.4g}')
    print(f'\n     tr(A) = {" + ".join(diag_parts)} = {trace:.6g}')
    print()
    divider('-', 45)
    print(f'  🎯 tr(A) = {trace:.6g}')
    # Verify with eigenvalues
    eigs = np.linalg.eigvals(A)
    eig_sum = np.sum(eigs.real)
    print(f'  🔁 Eigenvalue sum = {eig_sum:.4g}  (should match trace ✅)' if
          abs(trace - eig_sum) < 1e-6 else
          f'  Note: eigenvalue sum = {eig_sum:.4g}')

print('✅ All 9 operation functions defined.')
```

## 📘 Step 7: Demo — All Operations with Sample Matrices

This section demonstrates **all 9 operations automatically** using pre-defined matrices — useful for testing and report submission.

```python
# ── Sample matrices ────────────────────────────────────────────
A2 = np.array([[3., 1.],
               [2., 4.]])

B2 = np.array([[1., 2.],
               [3., 1.]])

A3 = np.array([[2., 1., 3.],
               [0., 4., 1.],
               [5., 2., 1.]])

A_rect = np.array([[1., 2., 3.],
                   [4., 5., 6.]])

B_rect = np.array([[7., 8.],
                   [9., 10.],
                   [11., 12.]])

print('✅ Sample matrices defined.')
format_matrix(A2, 'A2 (2×2)')
format_matrix(B2, 'B2 (2×2)')
format_matrix(A3, 'A3 (3×3)')
```

```python
# ── DEMO 1: Addition ────────────────────────────────────────────
section('➕  DEMO: Addition  A2 + B2')
C = np.add(A2, B2)
print('  🔍 Step-by-Step:')
for i in range(2):
    for j in range(2):
        print(f'     C[{i+1},{j+1}] = {A2[i,j]:.4g} + {B2[i,j]:.4g} = {C[i,j]:.4g}')
format_matrix(C, 'Result  A2 + B2')
```

```python
# ── DEMO 2: Subtraction ─────────────────────────────────────────
section('➖  DEMO: Subtraction  A2 − B2')
C = np.subtract(A2, B2)
print('  🔍 Step-by-Step:')
for i in range(2):
    for j in range(2):
        print(f'     C[{i+1},{j+1}] = {A2[i,j]:.4g} − {B2[i,j]:.4g} = {C[i,j]:.4g}')
format_matrix(C, 'Result  A2 − B2')
```

```python
# ── DEMO 3: Multiplication ──────────────────────────────────────
section('✖️  DEMO: Multiplication  A_rect × B_rect')
C = np.matmul(A_rect, B_rect)
print(f'  A_rect shape: {A_rect.shape}  ×  B_rect shape: {B_rect.shape}  →  Result: {C.shape}')
print('  🔍 Step-by-Step (dot product per cell):')
for i in range(A_rect.shape[0]):
    for j in range(B_rect.shape[1]):
        terms = ' + '.join(f'{A_rect[i,k]:.4g}×{B_rect[k,j]:.4g}' for k in range(A_rect.shape[1]))
        print(f'     C[{i+1},{j+1}] = {terms} = {C[i,j]:.4g}')
format_matrix(C, 'Result  A_rect × B_rect')
```

```python
# ── DEMO 4: Transpose ───────────────────────────────────────────
section('🔄  DEMO: Transpose  A3ᵀ')
T = A3.T
print(f'  Original: {A3.shape}  →  Transposed: {T.shape}')
print('  🔍 Step-by-Step (swap row↔col index):')
for i in range(A3.shape[0]):
    for j in range(A3.shape[1]):
        print(f'     Aᵀ[{j+1},{i+1}] ← A[{i+1},{j+1}] = {A3[i,j]:.4g}')
format_matrix(T, 'Result  A3ᵀ')
```

```python
# ── DEMO 5: Determinant ─────────────────────────────────────────
section('📐  DEMO: Determinant  det(A2) and det(A3)')
for mat, name in [(A2,'A2'),(A3,'A3')]:
    det = np.linalg.det(mat)
    r = mat.shape[0]
    print(f'\n  {name} ({r}×{r}):')
    if r == 2:
        a,b,c_,d = mat[0,0],mat[0,1],mat[1,0],mat[1,1]
        print(f'     det = (a×d)−(b×c) = ({a}×{d})−({b}×{c_}) = {a*d}−{b*c_} = {det:.4g}')
    else:
        print(f'     Using cofactor expansion along Row 1:')
        for j in range(3):
            minor = np.delete(np.delete(mat,0,axis=0),j,axis=1)
            cofac = ((-1)**j)*np.linalg.det(minor)
            print(f'     C[1,{j+1}] = {((-1)**j)}×det(M{j+1}) = {cofac:.4g}')
        print(f'     det(A3) = {det:.4g}')
    status = 'NON-SINGULAR ✅' if abs(det)>1e-10 else 'SINGULAR ❌'
    print(f'  🎯 det({name}) = {det:.4g}  →  {status}')
```

```python
# ── DEMO 6: Inverse ─────────────────────────────────────────────
section('🔃  DEMO: Inverse  A2⁻¹')
det = np.linalg.det(A2)
inv = np.linalg.inv(A2)
print(f'  det(A2) = {det:.4g}  →  Inverse exists ✅')
a,b,c_,d = A2[0,0],A2[0,1],A2[1,0],A2[1,1]
print(f'  🔍 2×2 Inverse formula: A⁻¹ = (1/det)×[[d,-b],[-c,a]]')
print(f'     = (1/{det:.4g}) × [[{d},{-b}],[{-c_},{a}]]')
format_matrix(inv, 'Result  A2⁻¹')
print('  🔁 Verification  A2 × A2⁻¹ =')
format_matrix(np.round(np.matmul(A2, inv),6), 'A2 × A2⁻¹ (Identity)')
```

```python
# ── DEMO 7: Eigenvalues ─────────────────────────────────────────
section('🌀  DEMO: Eigenvalues & Eigenvectors of A2')
evals, evecs = np.linalg.eig(A2)
print('  📖 Solve det(A2 − λI) = 0 for λ')
for i,(lam,vec) in enumerate(zip(evals, evecs.T)):
    print(f'\n  Eigenvalue  λ{i+1} = {lam.real:.4g}')
    print(f'  Eigenvector v{i+1} = [ {"  ".join(f"{v.real:.4g}" for v in vec)} ]')
    Av = np.matmul(A2, vec.real)
    lv = lam.real * vec.real
    print(f'  Verify A·v{i+1} ≈ λ{i+1}·v{i+1}: {"✅ PASS" if np.allclose(Av,lv,atol=1e-6) else "❌ FAIL"}')
```

```python
# ── DEMO 8: Rank ────────────────────────────────────────────────
section('📊  DEMO: Rank of A3 and A_rect')
for mat, name in [(A3,'A3'),(A_rect,'A_rect')]:
    rank = np.linalg.matrix_rank(mat)
    _, s, _ = np.linalg.svd(mat)
    r, c = mat.shape
    print(f'\n  {name} ({r}×{c}):')
    print(f'     Singular values: {[f"{sv:.4g}" for sv in s]}')
    print(f'     Non-zero count  : {np.sum(s > 1e-10)}')
    print(f'  🎯 rank({name}) = {rank}  (max possible = {min(r,c)})')
    print(f'     {"Full rank ✅" if rank == min(r,c) else "Rank-deficient ⚠️"}')
```

```python
# ── DEMO 9: Trace ───────────────────────────────────────────────
section('➰  DEMO: Trace of A2 and A3')
for mat, name in [(A2,'A2'),(A3,'A3')]:
    trace = np.trace(mat)
    r = mat.shape[0]
    diag_vals = [f'{mat[i,i]:.4g}' for i in range(r)]
    print(f'\n  {name} ({r}×{r}):')
    print(f'     Diagonal elements : {diag_vals}')
    print(f'     tr({name}) = {" + ".join(diag_vals)} = {trace:.4g}')
    eig_sum = np.sum(np.linalg.eigvals(mat).real)
    print(f'     Eigenvalue sum   = {eig_sum:.4g}  (matches trace: {"✅" if abs(trace-eig_sum)<1e-6 else "❌"})')
```
